### Imports and configs 

```python
import os 
import numpy as np 
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
DATA_PATHS = {
    "test_dir": "cnn_car_scratch_data/test",
    "model_path": "models/cnn_filter.h5"
}
```

### Load model and Test

```python
model = load_model(DATA_PATHS["model_path"])
```

```python
for i in os.listdir(DATA_PATHS["test_dir"]):
    img = image.load_img(os.path.join(DATA_PATHS["test_dir"], i), target_size=(64, 64))
    display(img)
    x = image.img_to_array(img)
    
    # Since the shape for testing should be 4 dimensions instead of 3 
    x = np.expand_dims(x, axis = 0)
    pred = model.predict(x)

    # Creating a list to present the output result as string
    isDamaged = ['Clear', 'Damage']
    prediction = isDamaged[np.argmax(pred)]
    print(prediction)
```
